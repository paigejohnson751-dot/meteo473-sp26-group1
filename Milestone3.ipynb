```python
#// ============================================================
#// script.js — JavaScript
#// ============================================================
#//
#// JavaScript runs in the browser and can read and modify the
#// HTML page after it loads. The browser's representation of
#// HTML as a tree of objects is called the DOM (Document Object
#// Model). Most JS interactions involve:
#//   1. Finding an element:  document.getElementById("someId")
#//   2. Listening for an event (click, change, submit)
#//   3. Updating the element's content, style, or attributes

#// Wait for the full HTML document to finish loading before
#// running any code. Without this, getElementById() might
#// return null because the element doesn't exist yet.
document.addEventListener("DOMContentLoaded", () => {

    #// ----------------------------------------------------------
    #// PLOT VIEWER
    #// Pattern for stepping through forecast-hour images.
    #// In your project, swap the placeholder <div> for an <img>
    #// and update img.src instead of plotLabel.textContent.
    #// ----------------------------------------------------------

    #// Define available forecast hours (like a Python list)
    const hours = ["000", "001", "002", "003", "004", "005", "006", "007", "008", "009", "010", "011", "012", "013", "014", "015", "016", "017", "018", "019", "020", "021", "022", "023", "024", "025", "026", "027", "028", "029", "030", "031", "032", "033", "034", "035", "036", "037", "038", "039"];
    let currentIndex = 0;   #// track the currently displayed hour

    #// Grab references to the HTML elements we need
    const hourSelect = document.getElementById("hourSelect");
    const plotLabel  = document.getElementById("plot-label");

    #// Helper function — update the display for a given index
    function showHour(index) {
        currentIndex = index;
        const hour = hours[index];

        #// Update the visible label in the placeholder box
        #plotLabel.textContent = `Showing: threat_${hour}.png`;
        document.getElementById("plotImage").src = `threat_{hour}.png`

        #// --- For your real project, replace the line above with: ---
        #// document.getElementById("plotImage").src = `images/threat_${hour}.png`;
        #// -----------------------------------------------------------

        #// Keep the dropdown in sync with what's currently shown
        hourSelect.value = hour;
    }

    #// When the dropdown value changes, find the matching index and show it
    hourSelect.addEventListener("change", () => {
        const index = hours.indexOf(hourSelect.value);
        showHour(index);
    });

    #// Previous button — step back one hour, wrapping at the beginning
    #// The % operator (modulo) makes the index wrap around, just like in Python
    document.getElementById("prevBtn").addEventListener("click", () => {
        const newIndex = (currentIndex - 1 + hours.length) % hours.length;
        showHour(newIndex);
    });

    #// Next button — step forward one hour, wrapping at the end
    document.getElementById("nextBtn").addEventListener("click", () => {
        const newIndex = (currentIndex + 1) % hours.length;
        showHour(newIndex);
    });

    #// ----------------------------------------------------------
    #// TOGGLE BUTTON
    #// classList.toggle() adds a CSS class if it's absent, removes
    #// it if present — toggling the element's visibility.
    #// ----------------------------------------------------------
    const toggleButton = document.getElementById("toggleButton");
    const toggleText   = document.getElementById("toggle-text");

    toggleButton.addEventListener("click", () => {
        #// toggle returns true if the class was ADDED (element is now hidden)
        const nowHidden = toggleText.classList.toggle("hidden");
        #// Update the button label to reflect the current state
        toggleButton.textContent = nowHidden ? "Show Details" : "Hide Details";
    });

    #// ----------------------------------------------------------
    #// CONTACT FORM
    #// event.preventDefault() stops the browser from reloading
    #// the page on submit (the default behavior for forms).
    #// ----------------------------------------------------------
    const contactForm    = document.getElementById("contactForm");
    const formResponse   = document.getElementById("form-response");

    contactForm.addEventListener("submit", (event) => {
        #// Prevent the default page-reload on form submission
        event.preventDefault();

        #// Read the value of the name input field
        #// .trim() removes leading/trailing whitespace (same as Python's str.strip())
        const name = document.getElementById("nameInput").value.trim();

        #// Update the response paragraph's text
        formResponse.textContent = name
            ? `Thanks, ${name}! Your comment was recorded.`
            : "Thanks for your comment!";

        #// Make the response visible by removing the "hidden" CSS class
        formResponse.classList.remove("hidden");

        #// Reset all form fields to their defaults
        contactForm.reset();
    });

});
```
